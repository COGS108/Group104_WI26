## Research Question

Has the general sentiment of mainstream U.S. pop music lyrics become more negative over the past 25 years (2000–2025)? 

More specifically, how did the sentiment of mainstream Spotify and Billboard pop music lyrics change during major societal crises (e.g., the COVID-19 pandemic) compared to the years immediately preceding those crises?

Independent variables: Time (year) and crisis period (pre-COVID vs during COVID).
Dependent variables: Lyrical sentiment scores and measures of lexical diversity derived from song lyrics.
Controls variable: include song popularity (Billboard rank, weeks on chart) and consistent sampling across years.

## Background and Prior Work


Over the past five decades, popular music lyrics have undergone systematic shifts in emotional tone, linguistic structure in lyrics, and thematic focus. (1) Lyrical analyses of song lyrics indicate that popular music has become increasingly repetitive and simple, with an emphasis on individual experiences and self expression. These major changes along with an increase in the negative emotional language indicated by words associated with sadness, anger, loneliness, and stress (2). Researchers suggest that these trends in lyrics are a reflection of a broader picture of cultural and psychological shifts within society, such as rising awareness of mental health issues and increased societal stress (3).

The quantitative studies using large lyrical datasets, for our project, the Billboard charting songs and long term collection of popular music demonstrated consistent increases in first person pronouns in songs and more emotional negativity words and phrases. These findings help us strengthen the idea and hypothesis that music lyrics serve as a cultural artifact that mirrors the people, their emotional states and the social priorities of the society as a whole. 

Moreover, prior research tells us that these long term trends are not uniform and they tend to shift during different periods of major disruption in society. For example, large-scale crises such the Covid 19 pandemic, the patterns in music  production, lyrical messages and listener preferences changed noticeably. Listeners often lean to music that can provide emotional comfort and relief from stress (4). 

We noticed that there are fewer studies that use consistent popularity-based datasets, like the U.S. Billboard Top 100, to directly compare crisis periods to immediately prior years, despite a wealth of evidence of long-term shifts in lyrical emotion. By investigating whether the COVID-19 pandemic reflects a visible shift from the long-term trend of rising negativity in popular U.S. pop music lyrics between 2000 and 2025, this project expands on previous research and continues to look at these patterns to find correlations between societal stress and music.

## Hypothesis


We, as a group, hypothesize that the period from 2000 to 2025, mainstream U.S. pop music lyrics have become increasingly negative in sentiment and less semantically diverse overall. We further hypothesize that the COVID-19 pandemic (2020–2021) is the reason for this temporarily disrupted this long-term trend, resulting in a relative decrease in negative sentiment or an increase in emotionally comforting lyrical content compared to the years immediately preceding the pandemic. 

According to our background research, there is a rise in negative emotional language and reduced lyrical complexity in popular music over history. However, many studies also show that during periods of societal stress and pressure, listeners tend to seek music that provides emotional relief or nostalgia to recall memories. Therefore, we expect the COVID-19 pandemic to produce a measurable deviation from the long-term trend of increasing negativity in the lyrics.
