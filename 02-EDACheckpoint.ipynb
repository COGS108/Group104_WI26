# Research Question

Has the general sentiment of mainstream U.S. pop music lyrics become more negative and repetitive due to a rise in mental health awareness over the past 25 years (2000–2025)? 

More specifically, how did the sentiment of mainstream Spotify and Billboard pop music lyrics change during major societal crises (e.g., the COVID-19 pandemic) compared to the years immediately preceding those crises?

Independent variables: Time (year) and crisis period (pre-COVID vs during COVID).
Dependent variables: Lyrical sentiment scores and measures of lexical diversity derived from song lyrics.
Controls variable: include song popularity (Billboard rank, weeks on chart) and consistent sampling across years.

## Background and Prior Work


Over the past five decades, popular music lyrics have undergone systematic shifts in emotional tone, linguistic structure in lyrics, and thematic focus. (1) Lyrical analyses of song lyrics indicate that popular music has become increasingly repetitive and simple, with an emphasis on individual experiences and self expression. These major changes along with an increase in the negative emotional language indicated by words associated with sadness, anger, loneliness, and stress (2). Researchers suggest that these trends in lyrics are a reflection of a broader picture of cultural and psychological shifts within society, such as rising awareness of mental health issues and increased societal stress (3).

The quantitative studies using large lyrical datasets, for our project, the Billboard charting songs and long term collection of popular music demonstrated consistent increases in first person pronouns in songs and more emotional negativity words and phrases. These findings help us strengthen the idea and hypothesis that music lyrics serve as a cultural artifact that mirrors the people, their emotional states and the social priorities of the society as a whole. 

Moreover, prior research tells us that these long term trends are not uniform and they tend to shift during different periods of major disruption in society. For example, large-scale crises such the Covid 19 pandemic, the patterns in music  production, lyrical messages and listener preferences changed noticeably. Listeners often lean to music that can provide emotional comfort and relief from stress (4). 

We noticed that there are fewer studies that use consistent popularity-based datasets, like the U.S. Billboard Top 100, to directly compare crisis periods to immediately prior years, despite a wealth of evidence of long-term shifts in lyrical emotion. By investigating whether the 9/11 attack and the COVID-19 pandemic reflects a visible shift from the long-term trend of rising negativity in popular U.S. pop music lyrics between 2000 and 2025, this project expands on previous research and continues to look at these patterns to find correlations between societal stress and music.

### A collection of lyrics from Billboard Top 100 songs ranging from 1946 to 2022
    This dataset contains five metrics: song name, lyrics, artist name, ranking year, and rank. Song name is the name of the song. Lyrics are the words of the song. The artist's name is the name of the artist who created the song being ranked. Ranking year is the year in which the song is being ranked. And rank is the song's placement in the top 100 during a given year. There are a few shortcomings with this dataset. The first shortcoming is that it only goes up to the year 2022, so we will either need to find a secondary dataset or lower the range we will be observing to match this dataset. Secondly, there are some null values in the lyrics column that we will have to filter out.

```python
import pandas as pd
import ast
lyric_df = pd.read_csv('./data/00-raw/lyric_dataset.csv')
## Rename columns for simplicity
lyric_df.columns = ['song','artists','year','rank','lyrics']
lyric_df['artists'] = lyric_df['artists'].apply(ast.literal_eval)
lyric_df
```

```python
print(f'There are {lyric_df.shape[0]} rows in the dataframe, with {lyric_df.shape[1]} columns.')
print("Examining null values, if any exist in the dataset")
print(lyric_df.isna().any())
print("As shown, there are null values in the Lyric column")
null_lyrics = lyric_df[lyric_df['lyrics'].isna()]
display(null_lyrics)
```

There are about 847 songs in our dataset without lyrics, but we plan to drop those. This dataset specifically shows the commonality of several mental and substance use disorders that have been in the United States from 1990 to 2017. Similar to the dataset above, the values in this dataset are also expressed as percentage of the total population in America. With a value such as 6.7% for anxiety disorders means that approximately 7 out of every 100 people were estimated to have an anxiety disorder during that year.

There are several important metrics that stand out in this dataset. Anxiety disorders are consistently the most common condition, staying between about 6.5% and 7.0% of the population, indicating a  persistent mental health burden of society. Following that, depression is the next most common condition; it affects roughly 4.6–4.8% of the population across the period, meaning about 5 in 100 people or 1 in 20 individuals experience depressive disorders in a given year. Although these percentages may seem small, they represent millions of individuals when applied to the U.S. population. In contrast to those growths, schizophrenia and bipolar disorder affect less than 1% of people (approximately 0.65% and 0.33-0.34% respectively), but they are often severe and lifelong conditions that require ongoing care and further support.

One of the most surprising patterns in the dataset is the rise in drug use disorders, which increased from about 2.4% in 1990 to more than 3.4% by 2017. That change represents millions of additional people struggling with substance use, and it reflects a  broader national concerns such as the opioid crisis. Meanwhile, alcohol use disorders declined slightly in the 1990s (~2.2-2.0% from 1990-1999) and then stabilized (~2.03-2.04% from 2000-2017), suggesting that patterns of substance use may be shifting rather than simply increasing overall. Although eating disorders remain below 0.55% and show a small rise through the early 2000s (~0.46-0.51% from 1990-2000) before starting to stabilize (~0.5 from 2001-2017), this mental problem should still be considered serious due to their high health risks for the society.

To analyze these percentages, it is helpful to consider scale as well. For example, with a U.S. population of roughly 325 million in 2017, a 6.6% prevalence of anxiety disorders corresponds to over 21 million people. For these dataset,  small percentage changes can represent millions of individuals in the country. Unlike clinical measurements like blood glucose levels, these metrics do not have a safe or dangerous range where we can estimate; they can only represent approximation of how common each disorder is in the population. However, steady increases, as shown in substance use disorders, can be a warning sign that public health crises are growing and that new medical programs might be needed to minimize them.

There are several important limitations and potential sources of bias in this dataset. In particular, these numbers are only approximation based on surveys, medical records, and statistical models, not exact counts. Many people never receive a formal diagnosis due to stigma, lack of access to healthcare, or cultural barriers, which means the true number has the potential to be higher or lower. Changes in diagnostic criteria and increased awareness over time can also make it look like some disorders are becoming more common when, in fact, they are simply being recognized more widely. In addition, because the data is aggregated at the national level, it hides important differences between regions (such as different states), age groups, and communities that experience mental health challenges in very different ways.

Overall, this dataset provides a valuable view of mental health trends in the United States. This dataset can highlights the high and persistent prevalence of anxiety and depression, the relative stability of severe psychiatric disorders like schizophrenia and bipolar disorder, and the concerning rise in drug use disorders. Most importantly, the dataset should be interpreted cautiously due to potential diagnostic, reporting, and aggregation biases.

```python
bb_lyrics = pd.read_csv('data/02-processed/billboard_lyrics.csv')
alc_drug = pd.read_csv('data/02-processed/drug_abuse_clean.csv')
display(bb_lyrics.head())
display(alc_drug.head())
```

```python
# evaluating list objects as lists using ast
import ast

bb_lyrics['artists'] = bb_lyrics['artists'].apply(ast.literal_eval)
bb_lyrics['lyrics'] = bb_lyrics['lyrics'].apply(ast.literal_eval)
```

```python
# adding features for us to visualize
import nltk
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('stopwords')
nltk.download('vader_lexicon')
nltk.download('punkt')

stop_words = set(stopwords.words('english'))
sia = SentimentIntensityAnalyzer()

def process_lyrics(lyric_list):
    unique_all = len(set(lyric_list))
    filtered_tokens = [w for w in lyric_list if w not in stop_words]
    unique_no_stop = len(set(filtered_tokens))
    sentiment = sia.polarity_scores(' '.join(lyric_list))
    return pd.Series([unique_all, unique_no_stop, sentiment['compound']])

bb_lyrics[['unique_words', 'unique_no_stop', 'sentiment_score']] = bb_lyrics['lyrics'].apply(process_lyrics)
```

```python
group_year = bb_lyrics.groupby('year')[['unique_words','unique_no_stop','sentiment_score']].mean().reset_index()
group_year
```

```python
group_year.plot(x='year', y=['unique_words', 'unique_no_stop'], marker='o', figsize=(10, 5))
```

```python
group_year.plot(x='year', y='sentiment_score', marker='o', figsize=(10, 5))
```
