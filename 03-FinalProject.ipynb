## Abstract

`This project examines changes in music sentiment over the past few decades alongside trends in alcohol and drug use disorders. Initially, our team intended to analyze how shifts in lyrical sentiment and repetitiveness might relate to broader mental health trends. However, due to the lack of accessible datasets covering recent years for mental health indicators, we redirected our focus to alcohol and drug use disorders, which are frequently referenced in modern music. Our objective was to determine whether mentions of alcohol and drugs in song lyrics correlate with real-world trends in substance use disorders.`

`To conduct this analysis, we used a dataset containing top-ranking songs and their lyrics from the 1940s onward and paired it with a dataset of hospital admissions related to alcohol and drug use disorders beginning in the 1990s. Drug and alcohol mentions in song lyrics also increased over time, while substance use disorder rates rose significantly beginning in the mid-2000s. Correlation analysis showed moderate positive relationships between substance mentions in lyrics and real-world substance use trends, as well as strong correlations between alcohol and drug use disorders. Additionally, higher frequencies of substance-related language were associated with lower lyrical sentiment scores, indicating more negative emotional tone.`

`Overall, the results suggest that references to drugs and alcohol in popular music may partially reflect broader societal trends in substance use while also being associated with more negative lyrical sentiment.`

```python
import matplotlib.pyplot as plt

bb_lyrics = lyric_df.groupby('year')[['unique_words', 'unique_no_stop', 'sentiment_score','unique_relative','alc_mentions','drug_mentions']].agg(['mean', 'min', 'max'])
bb_lyrics.columns = ['_'.join(col).strip() for col in bb_lyrics.columns.values]
bb_lyrics = bb_lyrics.reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(bb_lyrics['year'], bb_lyrics['unique_words_mean'], 
         label='Avg Unique Words', color='blue', linewidth=2)
plt.fill_between(bb_lyrics['year'], 
                 bb_lyrics['unique_words_min'], 
                 bb_lyrics['unique_words_max'], 
                 color='blue', alpha=0.15, label='Words Range (Min-Max)')
plt.plot(bb_lyrics['year'], bb_lyrics['unique_no_stop_mean'], 
         label='Avg No Stop Words', color='green', linewidth=2)
plt.fill_between(bb_lyrics['year'], 
                 bb_lyrics['unique_no_stop_min'], 
                 bb_lyrics['unique_no_stop_max'], 
                 color='green', alpha=0.15)

plt.title('Vocabulary Complexity Over Time')
plt.xlabel('Year')
plt.ylabel('Word Count')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
bb_lyrics.plot(x='year', y=['unique_words_mean', 'unique_no_stop_mean'], marker='o', figsize=(10, 5))
plt.title('Mean Unique Words in Lyrics Over Time')
plt.xlabel('Year')
plt.ylabel('Unique Words in a Song\'s Lyrics')
plt.show()
```

For the 'Vocabulary Complexity Over Time' graph, the blue and green lines indicate the average number of unique words per song and the average number of unique words excluding common stop words such as “the”, “and”, “you”, and “to” respectively. 
Before the 1960s, lyrics were written simply and used fewer unique words. However, starting from the 1960s, the complexity of lyrical vocabulary increased gradually through the early 2000s, indicating that a diverse range of distinct and meaningful words were used in lyrics. The plot had its highest peak in the 2005s and then slightly declined but remained at a higher rate compared to earlier decades. Overall, the two lines move closely over time, showing that lyrics in modern songs tend to use more varied and meaningful language rather than simply the repetition of common words in the mid 20th century. 


The graph after shows how the variety of words used in Billboard song lyrics has changed over time. The blue and orange lines show the average number of different words and average number of different meaningful words used in songs each year respectively. Both lines share a similar pattern which increases gradually from the 1960s and rises incredibly in the 1990s and 2000s where the lyrical diversity reaches its highest peak. Although there is a slight decrease after the 2000s, the number of unique words remains higher than in earlier years which suggests that modern songs tend to use a more diverse vocabulary.
