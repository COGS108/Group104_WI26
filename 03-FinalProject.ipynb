## Abstract

This project examines changes in music sentiment over the past few decades alongside trends in alcohol and drug use disorders. Initially, our team intended to analyze how shifts in lyrical sentiment and repetitiveness might relate to broader mental health trends. However, due to the lack of accessible datasets covering recent years for mental health indicators, we redirected our focus to alcohol and drug use disorders, which are frequently referenced in modern music. Our objective was to determine whether mentions of alcohol and drugs in song lyrics correlate with real-world trends in substance use disorders.

To conduct this analysis, we used a dataset containing top-ranking songs and their lyrics from the 1940s onward and paired it with a dataset of hospital admissions related to alcohol and drug use disorders beginning in the 1990s. The datasets were cleaned by removing null values, generating lyric sentiment scores using the VADER sentiment analysis tool, and constructing a dictionary of terms associated with alcohol and drugs. While this dictionary mentioned many commonly used references, some mentions may have been missed due to slang or contextual language. Our findings revealed several interesting findings. The number of unique words in lyrics increased over time, but when normalized by song length, diversity in unique words declined. Drug and alcohol mentions in song lyrics also increased over time, while substance use disorder rates rose significantly beginning in the mid-2000s. Correlation analysis showed moderate positive relationships between substance mentions in lyrics and real-world substance use trends, as well as strong correlations between alcohol and drug use disorders. Additionally, higher frequencies of substance-related language were associated with lower lyrical sentiment scores, indicating more negative emotional tone.

Overall, the results suggest that references to drugs and alcohol in popular music may partially reflect broader societal trends in substance use while also being associated with more negative lyrical sentiment.

## Background and Prior Work

Over the past decade, researchers have begun to examine how references to substances in popular music relate to real-world patterns of substance use. Studies of mainstream songs show that mentions of alcohol and drugs tend to increase and decrease alongside national substance use trends, suggesting that music reflects what is happening in society (1). At the same time, research on individual listeners finds that people who listen to more songs with substance-related lyrics are also more likely to report higher substance use themselves (2).

Together, these findings suggest that music is not just reflecting behavior, but may also play a role in shaping it. Substance-related themes appear across many different types of music, meaning exposure is widespread and not limited to one genre (2). Researchers believe that repeated exposure to these messages can make substance use seem more normal or acceptable over time (3).

In addition, music and pop culture often present substance use in a way that can feel normalized or even appealing, especially to younger audiences. This kind of portrayal can influence how people think about the risks and acceptability of using substances (3). When combined with the idea that music reflects real-world trends (1), this suggests that music not only mirrors behavior but can also influence and shape it over time.

By focusing specifically on substance-related references in popular music and how they align with real-world patterns of substance use, this project builds on past research, aiming to better understand the relationship between cultural musical expression and public health. Examining how often drugs and alcohol appear in widely consumed songs alongside trends in substance use disorders can provide insight into how cultural exposure may relate to real-life behaviors and public health outcomes.

### Exploring the Lyric Dataset

We begin by examining the dataset that includes Billboard Year-end Hot 100 songs and their lyrics. From the data cleaning and wrangling step, we created a few additonal features that might be helpful to analyze the lyric data.

```python
import matplotlib.pyplot as plt

bb_lyrics = lyric_df.groupby('year')[['unique_words', 'unique_no_stop', 'sentiment_score','unique_relative','alc_mentions','drug_mentions']].agg(['mean', 'min', 'max'])
bb_lyrics.columns = ['_'.join(col).strip() for col in bb_lyrics.columns.values]
bb_lyrics = bb_lyrics.reset_index()
```

```python
import seaborn as sns

sns.set_theme(style="whitegrid")

palette = sns.color_palette("Set2", 10)
color_total = palette[0]
color_no_stop = palette[1]

plt.figure(figsize=(12, 6))
plt.plot(bb_lyrics['year'], bb_lyrics['unique_words_mean'], 
         label='Avg Unique Words', color=color_total, linewidth=2.5)
plt.fill_between(bb_lyrics['year'], 
                 bb_lyrics['unique_words_min'], 
                 bb_lyrics['unique_words_max'], 
                 color=color_total, alpha=0.15, label='Words Range (Min-Max)')
plt.plot(bb_lyrics['year'], bb_lyrics['unique_no_stop_mean'], 
         label='Avg No Stop Words', color=color_no_stop, linewidth=2.5)
plt.fill_between(bb_lyrics['year'], 
                 bb_lyrics['unique_no_stop_min'], 
                 bb_lyrics['unique_no_stop_max'], 
                 color=color_no_stop, alpha=0.15)

plt.title('Lyrical Complexity: Total vs. Substantive Vocabulary', fontsize=14, pad=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Word Count', fontsize=12)
plt.legend(frameon=True, loc='upper left')
```

```python
bb_lyrics.plot(x='year', 
               y=['unique_words_mean', 'unique_no_stop_mean'], 
               color=[color_total, color_no_stop],
               marker='o', 
               markersize=4,
               figsize=(10, 5),
               linewidth=2)

plt.title('Mean Vocabulary Trends Over Time', fontsize=14, pad=15)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Mean Unique Words')
plt.grid(True, alpha=0.3)
```

For the 'Vocabulary Complexity Over Time' graph, the blue and green lines indicate the average number of unique words per song and the average number of unique words excluding common stop words such as “the”, “and”, “you”, and “to” respectively.

However, starting from the 1960s, the complexity of lyrical vocabulary increased gradually through the early 2000s, indicating that a diverse range of distinct and meaningful words were used in lyrics. The plot had its highest peak in the 2005s and then slightly declined but remained at a higher rate compared to earlier decades. Overall, the two lines move closely over time, showing that lyrics in modern songs tend to use more varied and meaningful language rather than simply the repetition of common words in the mid 20th century. 


The graph after shows how the variety of words used in Billboard song lyrics has changed over time. Both lines share a similar pattern which increases gradually from the 1960s and rises incredibly in the 1990s and 2000s where the lyrical diversity reaches its highest peak. #### Exploring our Alcohol and Drug Abuse Dataset, Merging With Lyrics Dataset

We now visualize the alcohol and drug dataframe and see if there are any interesting patterns within it and if any relationships exist between the content of the Billboard Hot 100 songs and the prevalence of alcohol or drug abuse in the United States at a given time.

```python
alc_pivot = drug_abuse.pivot(index='year', columns='cause', values='mean_val').reset_index()
merged = bb_lyrics.merge(alc_pivot, on='year')
merged.rename(columns={
    'Alcohol use disorders': 'alcohol',
    'Drug use disorders': 'drugs'
}, inplace=True)
merged.head()
```

We connect the Billboard Hot 100 lyrics dataset with the drug problem dataset to examine if lyrical trends in popular music align with more general societal patterns in substance usage. Song lyrics and information for songs that were listed on the Billboard Hot 100 chart between 1946 and 2022 are included in the lyrics dataset, which was acquired from Kaggle. We concentrate on the subset of songs that coincide with the available years in the substance disorder dataset for the purposes of our research.

A number of variables in the lyrics dataset capture the linguistic and thematic elements of each song. In addition to processed textual elements taken from the lyrics, these include the song title, artist, year, and chart position. 
These characteristics include the total number of words, the number of unique words, the sentiment score, the normalized lexical diversity, and the number of mentions of alcohol and drugs that were found using keyword matching. Before analysis, the lyrics were also simplified to standardize word forms. 

The variable 'year' is used to combine the two datasets so that lyrical characteristics and societal changes can be compared. This guarantees that every song in the lyrics collection is linked to the matching annual statistics for drug and alcohol use problems in the US. Each row of the combined dataset produced by the merge process reflects a song and the drug use statistics for the year the song debuted on the Billboard chart. 
More than 6,000 songs make up the entire collection, which also includes signs of societal substance usage and lyrical characteristics. This combined dataset enables us to investigate possible connections between broader societal trends in substance use and lyrical topics, such as references to drugs or alcohol and emotional feeling in lyrics. By combining these datasets, the research can investigate if changes in mainstream music's lyrical content are correlated with an increase in drug or alcohol use disorders. 


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
merged['drug_and_alc'] = merged['alcohol'] + merged['drugs']
scale_cols = ['sentiment_score_mean','drug_and_alc', 'alcohol', 'drugs', 'alc_mentions_mean','drug_mentions_mean']
merged[scale_cols] = scaler.fit_transform(
    merged[scale_cols])

color_health = palette[3]

plt.figure(figsize=(12, 6))

sns.lineplot(data=merged, x='year', y='sentiment_score_mean', 
             label='Lyrical Sentiment (VADER)', color=color_sentiment, 
             linewidth=3, marker='o', markersize=4)

sns.lineplot(data=merged, x='year', y='drug_and_alc', 
             label='Substance Use Disorders (Health Data)', color=color_health, 
             linewidth=3, marker='s', markersize=4)
plt.axvspan(2010, 2022, color='gray', alpha=0.1, label='Period of Highest Divergence')

plt.title('The Paradox of Modern Pop: Declining Sentiment vs. Rising Substance Abuse', 
          fontsize=15, pad=20)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Relative Intensity (Min-Max Normalized)', fontsize=12)

plt.legend(frameon=True, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)
plt.tight_layout()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=merged, x='year', y='drug_and_alc', 
             label='Total Substance Abuse Rates (Health Data)', 
             color=color_health)
sns.lineplot(data=merged, x='year', y='drug_mentions_mean', 
             label='Lyrical Drug Mentions', 
             color=color_drug)

sns.lineplot(data=merged, x='year', y='alc_mentions_mean', 
             label='Lyrical Alcohol Mentions', 
             color=color_alc)
plt.axvspan(1995, 2022, color='gray', alpha=0.05, label='Era of Rising Correlation')
plt.title('Art Reflecting Life: The Synchronized Rise of Lyrical Mentions and Abuse Rates', 
          fontsize=15, pad=20, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Relative Intensity (Min-Max Normalized)', fontsize=12)
plt.legend(frameon=True, loc='upper left', fontsize=10)
plt.tight_layout()
```

While exploratory analysis is useful, we attempt to validate our results statistically using OLS regression to see if alcohol mentions in lyrics (among other features) are a strong predictor for alcohol abuse rates.

```python
import statsmodels.api as sm
import pandas as pd

X = merged[['alc_mentions_mean', 'sentiment_score_mean', 'year', 'unique_words_mean']]
X = sm.add_constant(X) 

y = merged['alcohol']

model = sm.OLS(y, X).fit()
print(model.summary())
```

Based on the OLS regression, we can conclude that alcohol mentions in lyrics are not a direct, significant driver of alcohol abuse trends when controlling for time and song complexity. Instead, alcohol abuse appears to be driven by a long-term upward trend over the years, while the "simplification" of lyrics (lower unique word counts) tracks more closely with rising abuse than the specific mention of alcohol itself.

We then repeat the process to examine drug mentions

```python
X = merged[['drug_mentions_mean', 'sentiment_score_mean', 'year', 'unique_words_mean']]
X = sm.add_constant(X) 

y = merged['drugs']

model = sm.OLS(y, X).fit()
print(model.summary())
```

The OLS regression model for drug abuse rates demonstrates an exceptionally high goodness of fit, with an $R^2$ value of 0.946, indicating that approximately 95% of the variance in drug abuse prevalence is explained by the model's predictors. However, despite this high overall accuracy, the primary research hypothesis is not supported: `drug_mentions_mean` is statistically insignificant (p=0.995), meaning that the frequency of drug references in lyrics does not reliably predict drug abuse rates when other factors are considered. Similar to the alcohol model, the results are almost entirely dominated by the year variable (p<0.001), which shows a significant positive coefficient (0.0308), highlighting a consistent upward trend in drug abuse over time that exists independently of lyrical content. Additionally, the low Durbin-Watson statistic (0.214) suggests strong positive autocorrelation, reinforcing that while the rise in drug abuse is highly linear over time, it cannot be statistically attributed to the shift in musical sentiment or substance-related vocabulary within this multivariate framework.

We now try to implement an interrupted time series to see if a given event "interrupts" the metric's existing trajectory. Specifically, we will examine the effects of the notable events displayed above and song lyric sentiment. We begin by defining a function that allows us to observe the significance:

```python
import statsmodels.formula.api as smf

def examine_year(year, event=""):
    """Performs an Interrupted Time-Series (ITS) regression on lyric sentiment.

    This function constructs a quasi-experimental ITS model to evaluate whether 
    a specific historical event caused a 'step change' or a 'trend change' 
    in the average sentiment of song lyrics.

    Args:
        year (int): The specific calendar year the event occurred (e.g., 2008).
        event (str, optional): A label for the event used to name the 
            internal dataframe columns and format results. Defaults to "".

    Returns:
        statsmodels.regression.linear_model.RegressionResultsWrapper: A fitted 
            OLS model object containing coefficients, p-values, and 
            goodness-of-fit metrics.

    Note:
        This function assumes a global 'merged' DataFrame exists containing 
        'year' and 'sentiment_score_mean' columns.
    """
    its_df = merged.copy()

    its_df['time'] = its_df['year'] - its_df['year'].min()
    its_df[f'post_{event}'] = (its_df['year'] >= year).astype(int)
    its_df[f'time_after_{event}'] = (its_df['year'] - year).clip(lower=0)

    model_its = smf.ols(f'sentiment_score_mean ~ time + post_{event} + time_after_{event}', data=its_df).fit()
    print(model_its.summary())
    return model_its

examine_year(2020, 'covid')
```

As shown, the `post_covid` variable shows that sentiment decreases very slightly after 2020, but the p-value of 0.77 shows that the result is not significant.

```python
examine_year(2008, 'recession')
```

The Interrupted Time Series (ITS) analysis of the 2008 recession reveals a significant and somewhat counter-intuitive shift in musical sentiment. While the model confirms a strong long-term negative trend in sentiment (p=0.001), where scores decreased by approximately 0.025 units per year, there was a statistically significant positive "step change" immediately following the recession (coef=0.2454, p=0.030). This suggests that in the immediate wake of the economic crisis, mainstream song lyrics actually became more positive, potentially serving as a form of "escapism" or "comfort" for listeners. However, this shift did not fundamentally alter the trajectory of music; the slope change (p=0.486) was not significant, indicating that after the initial boost, sentiment simply returned to its gradual, decades-long decline. With an $R^2$ of 0.602, the model effectively demonstrates that while the 2008 recession caused a temporary positive spike in the "mood" of lyrics, it was not enough to reverse the broader cultural trend toward darker lyrical content.
